```python
#importing pandas library 
import pandas as pd
```

```python
#reading the csv file 
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
#data information 
df.info()
```

```python
#getting statistical information on numerical columns
df.describe()
```

```python
#finding null values 
df.isnull().sum()
```

```python
#imputing null values in Review Rating by assigning the median value for each category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
#Renaming the column names to snake case 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_') 
```

```python
#Renaming the purchase amount column
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Creating a new column Age Group  
list1 = ['Youth','Adult','Middle Adult','Senior Citizen']
df['age_group'] = pd.qcut(df['age'],q=4,labels = list1)
```

```python
df.head()
```

```python
#Creating new column purchase_frequency_days  
frequency_mapping ={
'Fortnightly': 14,
 'Weekly':7,
 'Monthly':30,
 'Quaterly':90,
 'Bi-Weekly':14,
 'Annually':365,
 'Every 3 months':90} 
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#coomparing discount_applied and promo_code_used is same 
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column as it is a duplicate column as discount_applied
df = df.drop('promo_code_used', axis=1)
```
